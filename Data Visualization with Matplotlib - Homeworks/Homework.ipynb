```python
# Import numpy and matplotlib

import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
x = np.arange(51) # Create an array from 0 to 50
y = np.arange(0,101,2) # Create an array from 0 to 100 with a step of 2
z = x ** 3 # Calculate the cube of each element in the array x
```

```python
fig1 = plt.figure() # Create a new figure

axes = fig1.add_axes([0,0,1,1]) # Add axes to the figure

axes.plot(x,y, color = "red") # Plot the x and y data with a red color
axes.set_title("X vs Y Graph") # Set the title of the graph
```

```python
fig2 = plt.figure(figsize = (8,6)) # Create a new figure with a specific size (8 inches by 6 inches)

axes1 = fig2.add_axes([0, 0.1, 0.8, 0.8]) # Add outer axes to the figure
axes1.plot(x,z, color = "purple") # Plot the x and z data with a purple color
axes1.set_title("Outer Graph - X vs Z Graph") # Set the title of the outer graph

axes2 = fig2.add_axes([0.1, 0.5, 0.4, 0.3]) # Add inner axes to the figure
axes2.plot(x,y, color = "red") # Plot the x and y data with a red color
axes2.set_title("Inner Graph - X vs Y Graph") # Set the title of the inner graph
```

```python
fig, axes = plt.subplots(nrows = 1, ncols = 2, figsize = (10,6)) # Create a figure with 1 row and 2 columns of subplots, with a specific size (10 inches by 6 inches)

axes[0].plot(x,y, color = "red", marker = "o", markerfacecolor = "black") # Plot x and y data with red color, circular markers, and black marker face color
axes[0].set_title("X vs Y Graph") # Set the title of the first subplot

axes[1].plot(x,z,color = "purple", linewidth = 3, linestyle = "-.") # Plot x and z data with purple color, line width of 3, and dash-dot linestyle
axes[1].set_title("X vs Z Graph") # Set the title of the second subplot

plt.tight_layout() # Adjust the spacing between subplots for better layout
```
